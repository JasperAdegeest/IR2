```python
import matplotlib.pyplot as plt
import numpy as np

totalclick, totalnone = [], []

with open("propfile.txt") as f:
    for line in f:
        try:
            [prop, click] = line.split();
            if int(click) == 0:
                totalclick.append(float(prop))
            else:
                totalnone.append(float(prop)*10)
        except:
            continue
```

```python
n, bins, patches = plt.hist(totalclick, 2000)
plt.axis([0,1,0,4000])
plt.grid(True)
plt.title("number of none clicks by propensity")
plt.xlabel('propensity')
plt.ylabel('number of occurences')
plt.savefig("noneClickByPropensity.png")
plt.show()
```

```python
n, bins, patches = plt.hist(totalnone, 5000)
plt.axis([0,1,0,40000])
plt.grid(True)
plt.title("number of clicks by propensity")
plt.xlabel('propensity')
plt.ylabel('number of occurences')
plt.savefig("clickByPropensity.png")
plt.show()
```

```python
print("number of not clicked: %i" %(len(totalnone)))
print("number of click: %i" % (len(totalclick)))
print("ratio: %.2f" % (len(totalnone)/len(totalclick)))
```

```python
print("average propensity not clicked: %.3f" % np.mean(totalnone))
print("average propensity click: %.3f" % np.mean(totalclick))
```
